# **Visualización de los datos**

La visualización de datos es una parte importante de la exploración de datos. Cuando se trata de librerías de visualización de datos, hay varias opciones basadas en Python para elegir, incluyendo [Matplotlib](https://matplotlib.org/), [Seaborn](https://seaborn.pydata.org/), [Plotly](https://plotly.com/), entre otras.

* **Matplotlib**: Es una biblioteca de trazado básica en Python que proporciona una amplia gama de funciones para crear visualizaciones estáticas y dinámicas. Es altamente personalizable y se puede utilizar para crear gráficos de líneas, gráficos de barras, gráficos de dispersión, gráficos de torta, diagramas de caja, entre otros.

* **Seaborn**: Es una biblioteca de visualización de datos basada en Matplotlib que proporciona una interfaz de alto nivel para crear gráficos atractivos y estadísticas descriptivas. Seaborn es especialmente útil para la visualización de datos estadísticos complejos y la creación de gráficos de correlación, distribución y regresión.

* **Plotly**: Es una biblioteca interactiva de visualización de datos que permite crear gráficos interactivos y visualizaciones web. Plotly ofrece una amplia gama de tipos de gráficos, incluyendo gráficos de barras, gráficos de líneas, gráficos de dispersión, gráficos de áreas, gráficos de torta y gráficos 3D. Además, permite crear visualizaciones interactivas con controles deslizantes, botones y ventanas emergentes.

* **Bokeh**: Es una biblioteca de visualización interactiva que se enfoca en la creación de gráficos interactivos y paneles de control para explorar y presentar datos. Bokeh permite crear gráficos interactivos en el navegador web con capacidades de zoom, pan y selección. También es compatible con la creación de paneles de control y aplicaciones interactivas utilizando Python o lenguajes front-end como JavaScript.

## **Biblioteca Matplotlib**

**plt** en **matplotlib.pyplot** es un alias comúnmente utilizado para referirse a la biblioteca **`matplotlib.pyplot`**. Aquí tienes algunos atributos y métodos comunes que puedes utilizar con **`plt`** para crear y personalizar gráficos:

Atributos:
* **plt.xlabel()**: Establece el texto para la etiqueta del eje x.
* **plt.ylabel()**: Establece el texto para la etiqueta del eje y.
* **plt.title()**: Establece el título del gráfico.
* **plt.legend()**: Muestra una leyenda para los elementos del gráfico.
* **plt.xticks()**: Establece las ubicaciones y etiquetas de las marcas del eje x.
* **plt.yticks()**: Establece las ubicaciones y etiquetas de las marcas del eje y.
* **plt.xlim()**: Establece los límites del eje x.
* **plt.ylim()**: Establece los límites del eje y.
* **plt.grid()**: Agrega una cuadrícula al gráfico.

Métodos:
* **plt.plot()**: Crea un gráfico de líneas o puntos.
* **plt.scatter()**: Crea un gráfico de dispersión.
* **plt.bar()**: Crea un gráfico de barras verticales.
* **plt.barh()**: Crea un gráfico de barras horizontales.
* **plt.pie()**: Crea un gráfico de torta.
* **plt.hist()**: Crea un histograma.
* **plt.boxplot()**: Crea un gráfico de caja.
* **plt.fill_between()**: Crea un gráfico de área.
* **plt.imshow()**: Muestra una imagen en el gráfico.
* **plt.show()**: Muestra el grafico en pantalla.

## **Ejemplo plt.plot()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.plot()** para personalizar el gráfico:

* **x:** Secuencia de valores que representan los datos del eje x.
* **y:** Secuencia de valores que representan los datos del eje y.
* **formato:** Cadena que especifica el formato del trazado. Por ejemplo, "b-" representa una línea azul, "ro" representa puntos rojos, etc.
* **color:** Color del trazado. Puede especificarse como una cadena de color ("blue", "red", "green", etc.) o mediante un código hexadecimal (#RRGGBB) o un código de color abreviado.
* **linewidth (o lw):** Ancho de línea.
* **linestyle (o ls):** Estilo de línea. Puede ser "-", "--", "-.", ":".
* **marker:** Marcador para los puntos en el gráfico. Puede ser "o", ".", "s", "*", etc.
* **markersize (o ms):** Tamaño de los marcadores.
* **label:** Etiqueta que se muestra en la leyenda del gráfico.
* **alpha:** Valor de opacidad del trazado. Un valor de 0.0 significa completamente transparente y 1.0 significa completamente opaco.

Mayor informacion de 
(Marker)[https://matplotlib.org/stable/api/markers_api.html])
(Linestyle)[https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html]

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange', 'watermelon']
counts = [40, 100, 30, 55, 75]
bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple']

ax.plot(fruits, counts, label=bar_labels, color=bar_colors[0], linewidth=1, linestyle='solid', marker='P')

ax.set_ylabel('fruit supply')
ax.set_xlabel('fruits')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

## **Ejemplo plt.scatter()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.scatter()** para personalizar el gráfico:

* **x:** Secuencia de valores que representan los datos del eje x.
* **y:** Secuencia de valores que representan los datos del eje y.
* **s:** Tamaño de los marcadores en el gráfico. Puede ser un solo valor numérico o una secuencia que especifique el tamaño de cada marcador individualmente.
* **c:** Color de los marcadores. Puede ser una cadena de color ("blue", "red", "green", etc.), una secuencia de colores que especifique el color de cada marcador individualmente, o un arreglo numérico que mapee los valores a colores utilizando un mapa de colores (colormap).
* **marker:** Marcador para los puntos en el gráfico. Puede ser "o", ".", "s", "*", etc.
* **alpha:** Valor de opacidad de los marcadores. Un valor de 0.0 significa completamente transparente y 1.0 significa completamente opaco.
* **cmap:** Mapa de colores utilizado para mapear los valores a colores cuando se utiliza el parámetro c.
* **label:** Etiqueta que se muestra en la leyenda del gráfico.
* **edgecolors:** Color del borde de los marcadores.
* **linewidths:** Ancho del borde de los marcadores.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange', 'watermelon']
counts = [40, 100, 30, 55, 75]
bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple']

ax.scatter(fruits, counts, 10, color=bar_colors, marker='P')

ax.set_ylabel('fruit supply')
ax.set_xlabel('fruits')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

## **Ejemplo plt.bar() - plt.barh()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.bar()** para personalizar el gráfico:

* **x:** Coordenadas de las barras en el eje x.
* **height:** Altura de las barras.
* **width:** Ancho de las barras. Por defecto es 0.8.
* **align:** Alineación de las barras. Puede ser 'center' (centradas), 'edge' (en el borde izquierdo) o 'tip' (en el extremo izquierdo).
* **color:** Color de las barras. Puede ser una cadena de color ("blue", "red", "green", etc.) o una secuencia de colores que especifique el color de cada barra individualmente.
* **edgecolor:** Color del borde de las barras.
* **linewidth:** Ancho del borde de las barras.
* **tick_label:** Etiquetas de las barras en el eje x.
* **label:** Etiqueta que se muestra en la leyenda del gráfico.
* **alpha:** Valor de opacidad de las barras. Un valor de 0.0 significa completamente transparente y 1.0 significa completamente opaco.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange', 'watermelon']
counts = [40, 100, 30, 55, 75]
bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
for i, count in enumerate(counts):
    ax.text(i, count, str(count), ha='center', va='bottom')

# ax.barh(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_xlabel('fruits')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

## **Ejemplo plt.pie()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.pie()** para personalizar el gráfico:

* **x:** Secuencia de valores numéricos que representan los tamaños relativos de cada categoría.
* **labels:** Etiquetas para cada categoría en la torta.
* **colors:** Colores de cada sector de la torta. Puede ser una cadena de color ("blue", "red", "green", etc.) o una secuencia de colores que especifique el color de cada sector individualmente.
* **explode:** Desplazamiento de cada sector de la torta desde el centro. Puede ser una secuencia de valores numéricos que indique el desplazamiento para cada sector individualmente.
* **autopct:** Formato para mostrar los porcentajes en cada sector. Puede ser una cadena de formato como "%1.1f%%" para mostrar un decimal con un dígito después del punto decimal.
* **shadow:** Booleano que indica si se muestra una sombra en la torta.
* **startangle:** Ángulo inicial de rotación de la torta.
* **counterclock:** Booleano que indica si los sectores se dibujan en sentido de las agujas del reloj (False) o en sentido contrario a las agujas del reloj (True).
* **wedgeprops:** Diccionario que permite personalizar las propiedades de los sectores de la torta, como el color del borde, el ancho del borde, etc.
* **textprops:** Diccionario que permite personalizar las propiedades del texto en los sectores, como el color, el tamaño de fuente, etc.
* **pctdistance:** Distancia del porcentaje desde el centro de la torta.
* **labeldistance:** Distancia de las etiquetas desde el centro de la torta.
* **radius:** Radio de la torta.
* **frame:** Booleano que indica si se muestra un marco alrededor de la torta.
center: Coordenadas del centro de la torta.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange', 'watermelon']
counts = [40, 100, 30, 55, 75]
bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple']

ax.pie(counts, labels=fruits, colors=bar_labels, autopct='%1.1f%%')

ax.set_title('Fruit supply by kind and color')

plt.show()
```

## **Ejemplo plt.hist()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.hist()** para personalizar el gráfico:

* **x:** Secuencia de valores numéricos que se utilizarán para crear el histograma.
* **bins:** Especifica cómo se deben agrupar los valores en el histograma. Puede ser un número entero que indica el número de contenedores/bins a utilizar o una secuencia que especifica los límites de los contenedores.
* **range:** Rango de valores que se incluirán en el histograma. Puede ser una tupla que especifique el valor mínimo y máximo del rango.
* **density:** Booleano que indica si se desea normalizar el histograma para obtener una distribución de densidad de probabilidad. Si se establece en True, el área total del histograma será igual a 1.
* **cumulative:** Booleano que indica si se desea mostrar el histograma acumulativo en lugar del histograma de frecuencia. Si se establece en True, cada bin mostrará la suma acumulativa de las frecuencias de todos los bins anteriores.
* **histtype:** Tipo de histograma a mostrar. Puede ser 'bar' (histograma de barras), 'barstacked' (histograma de barras apiladas), 'step' (histograma de línea sin relleno) o 'stepfilled' (histograma de línea con relleno).
color: Color de las barras del histograma.
* **edgecolor:** Color del borde de las barras del histograma.
linewidth: Ancho del borde de las barras del histograma.
* **alpha:** Valor de opacidad de las barras del histograma.
* **label:** Etiqueta que se muestra en la leyenda del gráfico.
orientation: Orientación del histograma. Puede ser 'vertical' (por defecto) o 'horizontal'.
* **rwidth:** Ancho relativo de las barras del histograma como un porcentaje del ancho total del bin.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

data = [78, 90, 80, 70, 99, 100, 67, 88, 95, 70, 78, 89, 84, 93, 73]

ax.hist(data, bins=4, edgecolor='black', color="green")

ax.set_xlabel('Valores')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma')

plt.show()
```

## **Ejemplo plt.boxplot()**

A continuación, se describen algunos de los parámetros más comunes que se pueden utilizar con **plt.boxplot()** para personalizar el gráfico:

* **x:** Datos numéricos o secuencia de datos numéricos que se utilizarán para crear el diagrama de caja.
* **labels:** Etiquetas que se mostrarán en el eje X para cada caja.
* **notch:** Booleano que indica si se debe dibujar una muesca en la caja para indicar la incertidumbre sobre la mediana.
bootstrap: Número entero que indica el número de veces que se realizará el método bootstrap para calcular los intervalos de confianza alrededor de la mediana.
* **vert:** Booleano que indica si las cajas se dibujarán verticalmente (True) o horizontalmente (False).
* **patch_artist:** Booleano que indica si se deben utilizar artistas de parche para dibujar las cajas.
* **meanline:** Booleano que indica si se debe dibujar una línea para indicar la media.
* **showmeans:** Booleano que indica si se deben mostrar los puntos que representan la media.
* **showfliers:** Booleano que indica si se deben mostrar los valores atípicos.
* **whiskers:** Float o número que controla la longitud de los bigotes en relación a los cuartiles.
* **widths:** Float o secuencia de floats que especifica el ancho de las cajas.
* **boxprops:** Diccionario de propiedades que se aplicarán a las cajas.
* **medianprops:** Diccionario de propiedades que se aplicarán a las líneas de la mediana.
* **whiskerprops:** Diccionario de propiedades que se aplicarán a los bigotes.
* **capprops:** Diccionario de propiedades que se aplicarán a las tapas de los bigotes.
* **flierprops:** Diccionario de propiedades que se aplicarán a los valores atípicos.
* **meanprops:** Diccionario de propiedades que se aplicarán a las líneas de la media.
* **showbox:** Booleano que indica si se deben mostrar las cajas.
* **showcaps:** Booleano que indica si se deben mostrar las tapas de los bigotes.
* **showmedians:** Booleano que indica si se deben mostrar las líneas de la mediana.
* **showwhiskers:** Booleano que indica si se deben mostrar los bigotes.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

data = [78, 90, 80, 70, 99, 100, 67, 88, 95, 70, 78, 89, 84, 93, 73]

ax.boxplot(data, vert=False)
# ax.boxplot(data, vert=False, notch=True, widths=0.5, patch_artist=True, boxprops={'facecolor': 'lightblue', 'linewidth': 2},
#            medianprops={'color': 'red', 'linewidth': 2}, whiskerprops={'color': 'black', 'linewidth': 2},
#            capprops={'color': 'black', 'linewidth': 2}, flierprops={'marker': 'o', 'markerfacecolor': 'green', 'markeredgecolor': 'black'})

ax.set_xlabel('Valores')
ax.set_ylabel('Datos')
ax.set_title('Diagrama de Cajas')

plt.show()
```
