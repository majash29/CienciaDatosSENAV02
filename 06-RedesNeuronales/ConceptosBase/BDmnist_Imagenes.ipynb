# **MNIST**

La base de datos **MNIST** (por sus siglas Instituto Nacional de Estándares y Tecnología), es una extensa colección de base de datos que se utiliza ampliamente para el entrenamiento de diversos sistemas de procesamiento de imágenes. También se emplea con frecuencia en el campo del aprendizaje automático para tareas de entrenamiento y prueba.

La base de datos MNIST consta de 60.000 imágenes de entrenamiento y 10.000 imágenes de prueba.

<img alt="Colaboratory logo" height="200px" src="https://upload.wikimedia.org/wikipedia/commons/f/f7/MnistExamplesModified.png" align="left" hspace="10px" vspace="0px">

```python
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos MNIST
mnist = keras.datasets.mnist

# Dividir el conjunto de datos en datos de entrenamiento y prueba
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Definir una muestra específica (en este caso, la imagen 4567)
sample_index = 4567

# Visualizar una muestra específica (en este caso, la imagen 4567)
if 0 <= sample_index < len(train_images):
  plt.imshow(train_images[sample_index], cmap='gray')
  plt.title(f"Etiqueta entrenamiento: {train_labels[sample_index]}")
  plt.axis('off')
  plt.show()

  plt.imshow(test_images[sample_index], cmap='gray')
  plt.title(f"Etiqueta testing: {test_labels[sample_index]}")
  plt.axis('off')
  plt.show()
else:
    print("El índice está fuera de rango.")
```
